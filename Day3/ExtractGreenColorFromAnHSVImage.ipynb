## Question 4: Extract green color from an HSV image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
filter  = cv2.imread('sunflower.jpg')
filter=cv2.cvtColor(filter,cv2.COLOR_BGR2RGB)
print(filter.shape)
plt.imshow(filter)
plt.show()
#cv2.imshow('green',filter)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
# It converts the BGR color space of image to HSV color space 
hsv = cv2.cvtColor(filter, cv2.COLOR_BGR2HSV) 
    
# Threshold of green in HSV space 
lower_green = np.array([36, 25, 25])         
upper_green = np.array([86, 255, 255])      
 # preparing the mask to overlay 
mask = cv2.inRange(hsv, lower_green, upper_green) 
 
# The black region in the mask has the value of 0, 
# so when multiplied with original image removes all non-green regions 
result = cv2.bitwise_and(filter, filter, mask = mask) 
```

```python
plt.imshow(mask)
plt.show()
```

```python
plt.imshow(result)
plt.show()
```
