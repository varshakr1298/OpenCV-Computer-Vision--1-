## Question 2: Click the photograph of your hand and draw and count the contours on it.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Loading the image of hand and converting it into grayscale
hand=cv2.imread('hand.png')
hand=cv2.resize(hand,(300,300))
handEdged=hand.copy()
#Original Image
hand=cv2.cvtColor(hand,cv2.COLOR_BGR2RGB)
plt.imshow(hand)
plt.show()
gray=cv2.cvtColor(hand,cv2.COLOR_BGR2GRAY)
gray=cv2.resize(gray,(300,300))
#cv2.imshow('hand',hand)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(gray)
plt.show()
```

```python
#Finding edges using Canny Edge Detection Algorithm
blurred = cv2.GaussianBlur(gray, (11, 11), 0)  #  Blurring the Image
edged = cv2.Canny(blurred, 70, 110) 
plt.imshow(edged)
plt.show()
```

```python
# Drawing contours on external edges
contours, hierarchy = cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print('Numbers of contours are : ', len(contours))
cv2.drawContours(hand, contours, -1, (0, 255, 255), 3) 
print("I found {} curves on my hand".format(len(contours)))
plt.imshow(hand)
plt.show()
```
