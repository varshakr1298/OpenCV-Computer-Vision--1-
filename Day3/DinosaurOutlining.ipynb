## Question 1: Dinosaur Outlining

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#Loading the image and converting it into grayscale
dinosaur = cv2.imread('d.png')
original=cv2.cvtColor(dinosaur,cv2.COLOR_BGR2RGB)
#cv2.imshow('dinosaur',dinosaur)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(original)
plt.show()
```

```python
gray=cv2.cvtColor(dinosaur,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
```

```python
gaussian = cv2.GaussianBlur(gray.copy(), (7,7), 0)
edge = cv2.Canny(gaussian, 200, 350)
#cv2.imshow('Edged Dinosaur',edge)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(edge)
plt.show()
```
