## Question 2: Shifting the image down and right using translation

```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading the image
elephant=cv2.imread('elephant.jpg')
#cv2.imshow('elephant',elephant)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
elephant=cv2.cvtColor(elephant,cv2.COLOR_BGR2RGB)
plt.imshow(elephant)
plt.show()
```

### Translating the image downward

```python
#Easy method using imutils library function translate()
translateDown=imutils.translate(elephant,0,300)
plt.imshow(translateDown)
plt.show()
```

```python
print(elephant.shape)
```

```python
#Translating the image downward using warpAffine() function
(h,w)=elephant.shape[:2]
q_h=int(h/4) #Calculating Quarter Height
```

```python
#Making a translation matrix
t = np.float32([[1, 0, 0], [0, 1, q_h]])
print(t)
```

```python
translatedDownward = cv2.warpAffine(elephant, t, (w, h))
plt.imshow(translatedDownward)
plt.show()
```

### Translating Rightward

```python
#Easy method using imutils library function translate()
translateRight=imutils.translate(elephant,300,0)
plt.imshow(translateRight)
plt.show()
```

```python
#Translating the image downward using warpAffine() function
#Calculating quarter width
q_w=int(w/4)
#Making a translation matrix
t = np.float32([[1, 0, q_w], [0, 1, 0]])
print(t)
```

```python
translatedRightward = cv2.warpAffine(elephant, t, (w, h))
plt.imshow(translatedRightward)
plt.show()
```

### Translating Diagonally down and right

```python
translateDiagonally=imutils.translate(elephant,q_w,q_h)
plt.imshow(translateDiagonally)
plt.show()
```

### Translating Diagonally up and left

```python
translateDiagonallyUp=imutils.translate(elephant,-q_w,-q_h)
plt.imshow(translateDiagonallyUp)
plt.show()
```
