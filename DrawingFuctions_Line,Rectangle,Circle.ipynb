## Question:3 Create a white image using np.ones and draw a line, rectangle and circle on it.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Creating white image window

```python
#Creating the white image window where the drawings will be displayed
img = np.ones((400, 400, 3), dtype = 'uint8')*255
a = img.copy()
#cv2.imshow('img', img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(img)
plt.show()
```

### Drawing a Line on the white image window

```python
line = cv2.line(a, (50,50), (350,350), (255, 0, 255), 3)
#cv2.imshow('line', line)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(a)
plt.show
```

### Drawing a Rectangle on the white image window

```python
b=img.copy()
rectangle = cv2.rectangle(b,(50, 50), (350, 350), (255, 0, 255), 3)
#cv2.imshow('b', b)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(b)
plt.show()
```

### Drawing a Circle on the white image window

```python
c = img.copy()
# Calculating the center of the image so that we can position the center of the circle to draw the circle
(x, y) = (int(img.shape[1]/2), int(img.shape[0]/2))
circle = cv2.circle(c, (x,y), (160), (255, 0, 255), 3)
#cv2.imshow('c', c)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(c)
plt.show()
```

### Combined Figure

```python
(x, y) = (int(img.shape[1]/2), int(img.shape[0]/2))

d = img.copy()

line_1 = cv2.line(d, (50,50), (350,350), (255, 0, 255), 3)

line_ = cv2.line(d, (50,350), (350,50), (255, 0, 255), 3)

rectangle_1 = cv2.rectangle(d,(50, 50), (350, 350), (255, 0, 255), 3)

circle_ = cv2.circle(d, (x,y), (60), (255, 0, 255), -1)

circle_1 = cv2.circle(d, (x,y), (80), (255, 0, 255), 3)

circle_2 = cv2.circle(d, (x,y), (100), (255, 0, 255), 3)

circle_3 = cv2.circle(d, (x,y), (120), (255, 0, 255), 3)

circle_4 = cv2.circle(d, (x,y), (140), (255, 0, 255), 3)

plt.imshow(d)
plt.show()
```
