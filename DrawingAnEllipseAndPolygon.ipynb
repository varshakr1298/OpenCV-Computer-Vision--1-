## Question 4: Drawing ellipse and polygon

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((400, 400, 3), dtype = 'uint8')
#cv2.imshow('img', img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(img)
plt.show()
```

```python
e = img.copy()
(x, y) = (int(img.shape[1]/2), int(img.shape[0]/2)) 
ellipse = cv2.ellipse(e, (x,y), (150,50), 0, 0, 360, (0, 0, 255) , -1) 
#First argument is image window, second is the center coordinates of the window, third is the axes length, fourth is the angle, fifth is the start anlge, sixth is the end angle and seventh is color and 8th is thickness
#cv2.imshow('e', e)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(e)
plt.show()
```

### Drawing a polygon

```python
p=img.copy()
#Creating an array of five points
pts = np.array([[100,50],[200,300],[370,200],[150,100]], np.int32)
pts = pts.reshape((-1,1,2))
#Creating a polygon using polylines function where 1st argument is image window, 2nd is the array of points,3rd is boolean true for closed lines and fourth is color
polygon=cv2.polylines(p,[pts],True,(0,255,255),3)
#cv2.imshow('p',p)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(p)
plt.show()

#Drawing pentagon

img1 = np.zeros((300, 300, 3), dtype = "uint8")
penta = np.array([[[40,160],[120,100],[200,160],[160,240],[80,240]]], np.int32)
pentagon = cv2.polylines(img1, [penta], True, (255,120,255),3)
plt.imshow(img1)
plt.show()
```
