## Question1: Image Inpainting 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
flower=cv2.imread('sunflower.jpg')
#cv2.imshow('sunflower',flower)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(flower)
plt.show()
```

```python
mask=cv2.imread('mask.png',0)
#cv2.imshow('mask',mask)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(mask)
plt.show()
```

```python
# Performing Inpainting
ns = cv2.inpaint(flower, mask, 1, cv2.INPAINT_NS) 
telea = cv2.inpaint(flower, mask, 1, cv2.INPAINT_TELEA)
```

```python
ns=ns.copy()    
plt.imshow(telea)
plt.show()
```

```python
ns = cv2.inpaint(ns, mask, 1, cv2.INPAINT_NS) 
ns=ns.copy()  
ns = cv2.inpaint(ns, mask, 1, cv2.INPAINT_NS) 
ns=ns.copy() 
ns = cv2.inpaint(ns, mask, 1, cv2.INPAINT_NS) 
ns=ns.copy() 
ns = cv2.inpaint(ns, mask, 1, cv2.INPAINT_NS) 
#cv2.imshow('inpainted image',ns)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(ns)
plt.show()
cv2.imwrite('final impainted image.jpg',ns)
```
