## Full Body Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('people.jpg')
fullbody_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_fullbody.xml')
#cv2.imshow('smile',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(image)
plt.show()
```

```python
# Converting the image into grayscale for computation
import cv2
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
body = fullbody_cascade.detectMultiScale(gray, 1.03,3)
simage=image.copy()
for (bx,by,bw,bh) in body:
    cv2.rectangle(simage,(bx,by),((bx + bw),(by + bh)),(255,0,0),2)               
```

```python
plt.imshow(simage)
plt.show()
```
