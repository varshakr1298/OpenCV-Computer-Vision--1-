## Smile Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('image1.jpg')
smile_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_smile.xml')
#cv2.imshow('smile',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(image)
plt.show()
```

```python
# Converting the image into grayscale for computation
import cv2
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
smiles = smile_cascade.detectMultiScale(gray, 1.3,30)
simage=image.copy()
for (sx,sy,sw,sh) in smiles:
    cv2.rectangle(simage,(sx,sy),((sx + sw),(sy + sh)),(255,0,0),2)               
```
